```
# Ogata-Banks Solution
# v  : velocity
# L  : distance downstream of source
# t  : time
# D  : dispersion coefficient
# Co : source concentration


def conc(L,t,Co,v,D):
    import math
    conc=(Co/2.0)*(math.erfc((L - v*t)/(2.0*math.sqrt(D*t))+math.exp((v*L)/(D))*math.erfc((L + v*t)/(2.0*math.sqrt(D*t)))))
    return(conc)
